The main aim of this project is to predict the house price based on various
features which we will discuss as we go ahead

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

## for feature selection

from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

# to visiualise all the columns in the dataframe

pd.pandas.set_option('display.max_columns' , None)
```

```python
dataset = pd.read_csv('X_train.csv')
```

```python
dataset.head()
```

```python
# by Using the combination of Lasso model and SelectFeatureModel we are going to select the best features.
#
```

```python
features_nan = [i for i in dataset.columns if dataset[i].isnull().sum() > 1 and dataset[i].dtypes == 'O']

for i in features_nan:
    print("inside")
    print("{} : {} % Categorical missing values " .format(i, np.round(dataset[i].isnull().mean() , 4)))
```

```python
features_nan
```

```python

numerical_with_nan = [i for i in dataset.columns if dataset[i].isnull().sum() > 1 and dataset[i].dtypes != 'O']

for i in numerical_with_nan:
    print("{} : {} % numerical columns missing values " .format(i,np.around(dataset[i].isnull().mean(),4)))
```

```python

for i in numerical_with_nan:
    # we will replace nan values by using median, since there are outliers
    
    median_value = dataset[i].median()
    print("median value of " ,i, '  ', median_value)
    
    # create a new feature to capture nan values. here we are creating a new column with name as columnname+nan and 
    # replacing nan values with 1 and if not nan then it will be 0. After below code see the dataset and you will
    # find 3 new columns LotFrontage+nan, MasVnrArea+nan ,  GarageYrBltnan that will be having 1 for nan and o for not nan
    
    dataset[i+'nan'] = np.where(dataset[i].isnull(),1,0)
    
    # then in the original dataset fill the numerical caolumns that have nan value with median value 
    
    dataset[i].fillna(median_value, inplace = True)
    
    # now nan values are replaced for all 3 numerical columns, just cross verify once
    
dataset[numerical_with_nan].isnull().sum()
```

```python
# capture the dependent feature

y_train = dataset[['SalePrice']]
```

```python
## drop dependent feature from dataset, id is just a incremental value and is of
# no useful.

X_train = dataset.drop(['Id','SalePrice'],axis=1)
```

```python
### Apply Feature Selection
# first, I specify the Lasso Regression Model
# then select a suitable alpha(equivalent of  penalty). alpha value is very much important
# The bigger the alpha the less features that will be selected. So we need to select a very small alpha value.
# we can experiment with cross validation by using diff diff alpha value

# Then i use the selectFromModel object from sklearn, which will seelct the 
# features whose coefficients are non-zero

# selectfrom model, usually takes a algortihm which will actually helps you to reduce the 
# feature and Lasso Regression is one of them. this combination helps us to select the imp features for the model
#  remember the seed/random_state value, for test also we need to use the same seed value 

feature_sel_model = SelectFromModel(Lasso(alpha = 0.005, random_state=0))                                         
feature_sel_model.fit(X_train,y_train)
```

```python
feature_sel_model.get_support()  # count of all the independent featrures. true indicates imp featurea and that should be
# used
```

```python
# Lets print the number of total and selected features

# this is how we can make a list of the selected features

selected_feat = X_train.columns[(feature_sel_model.get_support())]

# lets print some stats

print('total features : {}' .format((X_train.shape[1])))
print('selected features : {}' .format(len(selected_feat)))

print("features with coefficents shark to zero : {}" .format(np.sum(feature_sel_model.estimator_.coef_==0)))
```
